```python
from selenium import webdriver  
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec

# to get the scores and hyperlinks of boxscore from footballdb.com
def get_week_score_links(url,week_k,game_num):
    driver=webdriver.Chrome(executable_path='\\Users\\codym\\OneDrive\\Desktop\\Accounting\\Data Scientist\\web-scrapping & APIs\\chromedriver')
    driver.get(url)
    score=[]
    links=[] 
    for i in range(1,game_num+1):
        for j in [3,5]:
            xpath='//*[@id="leftcol"]/table['+str(week_k)+']/tbody/tr['+str(i)+']/td['+str(j)+']'          
            wait=WebDriverWait(driver,10)
            loadxp='//*[@id="outbrain_widget_0"]/div/div[1]/div/span'
            wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
            element= driver.find_elements_by_xpath(xpath)
            score.append(element[0].text) 
        
        xpath='//*[@id="leftcol"]/table['+str(week_k)+']/tbody/tr['+str(i)+']/td[7]/a'          
        wait=WebDriverWait(driver,10)
        loadxp='//*[@id="outbrain_widget_0"]/div/div[1]/div/span'
        wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
        element= driver.find_elements_by_xpath(xpath)
        links.append(element[0].get_attribute('href'))  

    driver.close()
    score=[int(i) for i in score]
    return score,links
```

```python
# to get rush yard/attempt & interception & penalty time and yard & fumble and fumble lost
# of each game from footballdb.com
def get_rush_inn_pen_fum(rush,inn,pen,fum,fum_lost):
    global pen_yard
    for link in links:
        driver=webdriver.Chrome(executable_path='\\Users\\codym\\OneDrive\\Desktop\\Accounting\\Data Scientist\\web-scrapping & APIs\\chromedriver')
        driver.get(link)        
        for i in range(2,4):
            xpath='//*[@id="divBox_team"]/div[2]/div/table/tbody/tr[8]/td['+str(i)+']'            
            wait=WebDriverWait(driver,10)
            loadxp='//*[@id="divBox_team"]/div[3]/div/table/tbody/tr[8]/td[1]/b'
            wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
            element= driver.find_elements_by_xpath(xpath)
            rush.append(element[0].text)  
            
            xpath='//*[@id="divBox_team"]/div[3]/div/table/tbody/tr[5]/td['+str(i)+']'            
            wait=WebDriverWait(driver,10)
            loadxp='//*[@id="divBox_team"]/div[3]/div/table/tbody/tr[8]/td[1]/b'
            wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
            element= driver.find_elements_by_xpath(xpath)            
            inn.append(element[0].text.split('-')[0]) 
            
            xpath='//*[@id="divBox_team"]/div[3]/div/table/tbody/tr[6]/td['+str(i)+']'            
            wait=WebDriverWait(driver,10)
            loadxp='//*[@id="divBox_team"]/div[3]/div/table/tbody/tr[8]/td[1]/b'
            wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
            element= driver.find_elements_by_xpath(xpath)            
            pen.append(element[0].text.split('-')[0]) 
            pen_yard.append(element[0].text.split('-')[-1])
            
            xpath='//*[@id="divBox_team"]/div[3]/div/table/tbody/tr[7]/td['+str(i)+']'            
            wait=WebDriverWait(driver,10)
            loadxp='//*[@id="divBox_team"]/div[3]/div/table/tbody/tr[8]/td[1]/b'
            wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
            element= driver.find_elements_by_xpath(xpath)            
            fum.append(element[0].text.split('-')[0]) 
            fum_lost.append(element[0].text.split('-')[-1])
        driver.close()    
        
        rush=[float(i) for i in rush]
        inn=[int(i) for i in inn]
        pen=[int(i) for i in pen]
        pen_yard=[int(i) for i in pen_yard]
        fum=[int(i) for i in fum]
        fum_lost=[int(i) for i in fum_lost]
        
    return rush,inn,pen,pen_yard,fum,fum_lost    
```

```python
def reset_everything():
    rush=[]
    inn=[]
    pen=[]
    pen_yard=[]
    fum=[]
    fum_lost=[]
    
    passing=[]
    receive=[]
    tot=[]
    sack=[]
    kick=[]
    kicker_ratio=[]
    return rush,inn,pen,pen_yard,fum,fum_lost,passing,receive,tot,sack,kick,kicker_ratio
```

```python
# to get passing rating & receive yard/attempt & total of tackles & sacks &
# kick points and kicker_raito of each game from espn.com
def get_passing_rec_tot_sack_kick_kicker_ratio(url):
    driver=webdriver.Chrome(executable_path='\\Users\\codym\\OneDrive\\Desktop\\Accounting\\Data Scientist\\web-scrapping & APIs\\chromedriver')
    driver.get(url)
    
    for i in [1,3]:
        xpath='//*[@id="gamepackage-box-score"]/article/header/div/button['+str(i)+']'  
        wait=WebDriverWait(driver,10)
        loadxp='//*[@id="gamepackage-passing"]'
        wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
        element= driver.find_elements_by_xpath(xpath)
        # click the button 'rams'
        for ele in element:
            ele.click() 

        element= driver.find_elements_by_id("gamepackage-passing")
        passing.append(element[0].text.split(' ')[-1])    
        element= driver.find_elements_by_id("gamepackage-receiving")
        receive.append(element[0].text.split(' ')[-4])    
        element= driver.find_elements_by_id("gamepackage-defensive")
        tot.append(element[0].text.split(' ')[-7])
        sack.append(element[0].text.split(' ')[-5])
        element= driver.find_elements_by_id("gamepackage-kicking")
        kick.append(element[0].text.split(' ')[-1])
        
        kicker_com_fg=int(element[0].text.split(' ')[-5].split('/')[0]) 
        kicker_att_fg=int(element[0].text.split(' ')[-5].split('/')[-1])
        kicker_com_xp=int(element[0].text.split(' ')[-2].split('/')[0])
        kicker_att_xp=int(element[0].text.split(' ')[-2].split('/')[-1]) 
        kicker_ratio.append((kicker_com_fg+kicker_com_xp)/(kicker_att_fg+kicker_att_xp))
    driver.close()
    return passing,receive,tot,sack,kick,kicker_ratio
```

```python
# (as for kicker_ratio, if a game scoerd 0, there will be a problem cause denominator would be 0)
# so this is a fixed funtion.
def reset_everything_with_0_scored(): 
    rush=[]
    inn=[]
    pen=[]
    pen_yard=[]
    fum=[]
    fum_lost=[]
    
    passing=[]
    receive=[]
    tot=[]
    sack=[]
    kick=[]
    return rush,inn,pen,pen_yard,fum,fum_lost,passing,receive,tot,sack,kick
```

```python
# just to get the kicker_ratio for the game from espn.com (if there is a game scored 0)
def get_kicker_ratio(url):
    driver=webdriver.Chrome(executable_path='\\Users\\codym\\OneDrive\\Desktop\\Accounting\\Data Scientist\\web-scrapping & APIs\\chromedriver')
    driver.get(url)
    
    for i in [1,3]:
        xpath='//*[@id="gamepackage-box-score"]/article/header/div/button['+str(i)+']'  
        wait=WebDriverWait(driver,10)
        loadxp='//*[@id="gamepackage-passing"]'
        wait.until(ec.visibility_of_element_located((By.XPATH, loadxp)))
        element= driver.find_elements_by_xpath(xpath)
        # click the button 'rams'
        for ele in element:
            ele.click() 
        element= driver.find_elements_by_id("gamepackage-kicking")
        kicker_com_fg=int(element[0].text.split(' ')[-5].split('/')[0]) 
        kicker_att_fg=int(element[0].text.split(' ')[-5].split('/')[-1])
        kicker_com_xp=int(element[0].text.split(' ')[-2].split('/')[0])
        kicker_att_xp=int(element[0].text.split(' ')[-2].split('/')[-1]) 
        kicker_ratio.append((kicker_com_fg+kicker_com_xp)/(kicker_att_fg+kicker_att_xp))
    driver.close()
    return kicker_ratio
```

```python
def get_a_df(div,home,passing,rush,receive,kick,fum,fum_lost,pen,pen_yard,inn,tot,sack,score,kicker_ratio):
    df=pd.DataFrame([])
    df['div?']=div
    df['home?']=home
    df['passing']=passing
    df['rush']=rush
    df['receive']=receive
    df['kick']=kick
    df['fumble']=fum
    df['fum_lost']=fum_lost
    df['pen']=pen
    df['pen_yard']=pen_yard
    df['inn']=inn
    df['tot']=tot
    df['sack']=sack
    df['score']=score
    df['kicker_ratio']=kicker_ratio
    
    sample_len=len(df['passing'])
    oppo_tot=[0]*sample_len
    oppo_sack=[0]*sample_len
    oppo_inn=[0]*sample_len
    oppo_pen=[0]*sample_len
    oppo_pen_yard=[0]*sample_len
    for i in range(0,sample_len,2):
        oppo_tot[i]=tot[i+1]
        oppo_tot[i+1]=tot[i]
        oppo_sack[i]=sack[i+1]
        oppo_sack[i+1]=sack[i]    
        oppo_inn[i]=inn[i+1]
        oppo_inn[i+1]=inn[i]
        oppo_pen[i]=pen[i+1]
        oppo_pen[i+1]=pen[i]
        oppo_pen_yard[i]=pen_yard[i+1]
        oppo_pen_yard[i+1]=pen_yard[i]    
    df['oppo_tot']=oppo_tot
    df['oppo_sack']=oppo_sack
    df['oppo_inn']=oppo_inn
    df['oppo_pen']=oppo_pen
    df['oppo_pen_yard']=oppo_pen_yard

    df['pen_dif']=df['pen']-df['oppo_pen']
    df['pen_yard_dif']=df['pen_yard']-df['oppo_pen_yard']
    df=df.drop(['tot','sack','inn','pen','oppo_pen','pen_yard','oppo_pen_yard'],axis=1)    
    return df
```

```python
# in case it returns str for no actions for kicking
def get_new_kick(kick):
    new_kick=[]
    for i in kick:
        if len(i)<=2:
            new_kick.append(int(i))
        else:
            new_kick.append(0)
    return new_kick
```
